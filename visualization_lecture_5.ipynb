## Line Graphs 

The line chart is one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers connected by straight lines. You can customize the shape, size, color, and other aesthetic elements of the lines and markers for better visual clarity.

Here's a Python list showing the yield of apples (tons per hectare) over six years in an imaginary country called Kanto.

```python
# Yield of apples (tons per hectare) over a six year period in Kanto region
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

We can visualize how the yield of apples changes over time using a line chart. To draw a line chart, we can use the `plt.plot()` function.

```python
plt.plot(yield_apples)
```

Calling the `plt.plot()` function draws the line chart as expected. It also returns a list of plots drawn `[<matplotlib.lines.Line2D at 0x7f85b1f96b50>]`, shown within the output. We can include a semicolon (`;`) at the end of the last statement in the cell to avoiding showing the output and display just the graph.

```python
plt.plot(yield_apples);
```

Let's enhance this plot step-by-step to make it more informative and beautiful. Starting with....

### Customizing the x-axis

The X-axis of the plot currently shows list element indexes $0$ to $5$. The plot would be more informative if we could display the year for which we're plotting the data. To do this, we need to create another list of years. Then add that arguement to the fct `plt.plot()`. 

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]

# now to plot both years and yield_apples
plt.plot(years, yield_apples);
```

### Axis Labels

We can add labels to the axes to show what each axis represents using the `plt.xlabel()` and `plt.ylabel()` methods.

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Apple Yield (tons per hectare)');
```

### Plotting Multiple Lines

You can invoke the `plt.plot()` function once for each line to plot multiple lines in the same graph. Let's compare the yields of apples vs. oranges in Kanto.

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

### Chart Title and Legend

To differentiate between multiple lines, we can _include a **legend**_ within the graph using the `plt.legend()` function. We can also _set a **title**_ for the chart using the `plt.title()` function.

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.title('Crop Yield in Kanto')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apples', 'Oranges']);
```

### Line Markers

We can also _show **markers** for the data points on each line_ using the marker argument of `plt.plot()`. Matplotlib provides many different markers, like a circle, cross, square, diamond, etc. You can find the full list of marker types [here](https://matplotlib.org/3.1.1/api/markers_api.html).

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')
# labels
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
# title+legend
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

### Styling Lines and Markers

The `plt.plot()` function supports many arguments for styling lines and markers:

- `color` or `c`: Set the color of the line (supported colors)
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot

Check out the documentation for `plt.plot()` to learn [more](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html).

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
    # square markers, blue, solid line, 2 wide, 8 size markers, 2 edge width, navy edge color
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)
    # circle markers, red, dashed line, 3 width, 10 size markers, half opacity

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

The `fmt` argument provides a shorthand for specifying the _marker shape_, _line style_, and _line color_. It can be provided as the third argument to `plt.plot()`.
> fmt = '[marker][line][color]'

Note: _If you don't specify a line style in `fmt`, only markers are drawn. But you can include maker shape and line style only (i.e `plt.plot(years, apples, 's--')`) OR line style and line color only (i.e `plt.plot(years, apples, '--r')`)._

```python
plt.plot(years, apples, 's-b')
    # square markers, solid line, blue
plt.plot(years, oranges, 'o--r')
    # circle markers, dashed line, red

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

If you don't specify a line style in `fmt`, only markers are drawn...

```python
plt.plot(years, oranges, '--r') # line style and color WORK
plt.plot(years, apples, 's--') # marker shape and line style WORK
# plt.plot(years, apples, 's--') # marker shape and lin ecolor DONT WORK (no line will show)
plt.title("Yield of Oranges (tons per hectare)");
```

## Scatter Plot

In a scatter plot, the values of _2 variables are plotted as points_ on a 2-dimensional grid. Additionally, you can also _use a third variable to determine the size or color_ of the points. Let's try out an example.

The [Iris flower dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set) provides sample measurements of sepals and petals for three species of flowers. The Iris dataset is included with the Seaborn library (many more can be found with a quick google search) and can be loaded as a Pandas data frame.

```python
# Loading the dataset (into a Pandas dataframe)
iris_df = sns.load_dataset('iris')

# lets look at the dataset
iris_df
```

```python
# Let's see the unique species we're working with... 
iris_df.species.unique()
```

Let's try to **visualize the relationship between sepal length and sepal width**. The first thing we may do is create a line chart using `plt.plot()`...

```python
plt.plot(iris_df.sepal_length, iris_df.sepal_width);
```

The output graph is not very helpful or informative as there are too many combinations of the two properties within the dataset. There's no obvious simple relationship between them.

*So*, let's use **use a scatter plot to visualize how sepal length & sepal width vary** using the scatterplot function from the seaborn module (imported as sns).

```python
sns.scatterplot(x=iris_df.sepal_length, y=iris_df.sepal_width);

# the same can be achieved using matplotlib's pyplot -> plt.scatter()
```

### Adding Hues

As we can see in the above plot, it's still difficult to get info out of this. We can _color the dots using the flower species_ as a `hue`. We can also _make the points larger_ using the `s` argument.

```python
sns.scatterplot(x = iris_df.sepal_length, 
                y = iris_df.sepal_width, 
                hue = iris_df.species, # <=== color of points
                s = 75); # <=== size of points
```

Adding hues makes the plot much more informative. We can immediately tell that Setosa flowers have a smaller sepal length but higher sepal widths. In contrast, the opposite is true for Virginica flowers.

### Customizing Seaborn Figures 

Since Seaborn uses Matplotlib's plotting functions internally, we can use functions like `plt.figure()` and `plt.title()` to modify the figure.

```python
# customizations
plt.figure(figsize = (12, 6))
plt.title("Iris Sepal Dimensions")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

# plot
sns.scatterplot(x = iris_df.sepal_length, 
                y = iris_df.sepal_width, 
                hue = iris_df.species, 
                s = 75);
```

### Plotting using Pandas Data Frames

Seaborn has in-built support for Pandas data frames. Instead of passing each column as a series (i.e. `iris_df.species`), you can just provide column names and use the `data` argument to specify a data frame.

```python
sns.scatterplot(data = iris_df, 
                x = 'sepal_length', 
                y = 'sepal_width', 
                hue = 'species', 
                s = 75);
```

## Histogram

A histogram represents the distribution of a variable by creating bins (interval) along the range of values and using vertical bars to indicate the number of observations in each bin.

For example, let's __visualize the distribution of values of sepal width__ in the flowers dataset. We can use the `plt.hist()` function to create a histogram.

```python
# We have already loaded the 'iris' data into a Pandas df. 
# So know lets look into the `sepal_width` variable...
iris_df.sepal_width
```

```python
# Now lets visualize...
plt.title("Distribution of Sepal Width")
plt.hist(iris_df.sepal_width);

# the same can be achieved using seaborn -> sns.histplot() <- seems to have much more customizations
```

We can immediately see that the sepal widths lie in the range between $2.0$ and $4.5$, and a little over $35$ values are in the range 2.9 - 3.1 (though it is hard to be exact with these numbers as the ticks are not exact enough) which seems to be the most populous bin.

We can also look at this data as a frequency distribution using Seaborn. In a frequency distro, the values on Y-axis are percentagess instead of counts.

```python
sns.distplot(iris_df.sepal_width);
```

### Controlling the size and number of bins

We can control the number of bins or the size of each one using the `bins` argument.

```python
# Specifying the _number of bins__
plt.hist(iris_df.sepal_width, bins = 5);
```

```python
# Specifying the _boundaries (size) of each bin_
plt.hist(iris_df.sepal_width, 
         bins=np.arange(2, 5, 0.25));
```

```python
# Creating _bins of unequal sizes_
plt.hist(iris_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

### Multiple Histograms

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity so that one histogram's bars don't hide the others'.

Let's **draw separate histograms for each species of flowers**.

```python
# First we must QUERY the data for each species

setosa_df = iris_df[iris_df.species == 'setosa']
versicolor_df = iris_df[iris_df.species == 'versicolor']
virginica_df = iris_df[iris_df.species == 'virginica']
```

Let's create two hsitograms by **OVERLAPPING** them. Lowering the opacity (`alpha`) is very important so we can actually see the overlapping.

```python
plt.hist(setosa_df.sepal_width, 
         alpha=0.4, 
         bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, 
         alpha=0.4, 
         bins=np.arange(2, 5, 0.25));
```

Now let's create multiple histograms by **STACKING** them.

```python
plt.title('Distribution of Iris Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

Without the `stacked=True` parameter, the multiple histograms will be presented **SIDE-BY-SIDE**. 

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25));

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

## Bar Chart

Bar charts are quite similar to line charts in the way that they show a sequence of values. However, a bar is shown for each value, rather than points connected by lines. We can use the `plt.bar()` function to draw a bar chart.

_Note:_ The differences between Histograms and Bar Plots...
- Histograms are used to show distributions of variables while bar charts are used to compare variables. 
- Histograms plot binned quantitative data while bar charts plot categorical data. 
- Bars can be reordered in bar charts but not in histograms.

```python
# Variables
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]

# Plot
plt.bar(years, oranges);
```

Like histograms, we can stack bars on top of one another. We use the `bottom` argument of `plt.bar()` to achieve this.

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);

# NOTE
# The first line `plt.bar(years, apples)` is necessary to include. See why for yourself
```

### Bar Plots with Averages

Let's load and explore another sample dataset included with Seaborn, called `tips`. This dataset contains information about the sex, time of day, total bill, and tip amount for customers visiting a restaurant over the course of a week.

```python
tips_df = sns.load_dataset('tips')

tips_df
```

We might want to draw a bar chart to _visualize how the average bill amount varies across different days of the week_. One way to do this would be to compute the day-wise averages and then use `plt.bar`.

BUT, given that this is a very common use case, the Seaborn library provides a `barplot` function which can automatically compute averages.

```python
sns.barplot(x='day', y='total_bill', data=tips_df);
```

The lines cutting each bar represent the amount of variation in the values. It seems like the variation in the `total_bill` is relatively high on Fridays and low on Sunday.

We can also specify a hue argument to compare bar plots side-by-side based on a third feature, which in this case will be `sex`.

```python
sns.barplot(data = tips_df, 
            x = 'day', 
            y = 'total_bill', 
            hue = 'sex');
```

_Bonus:_ you can also make the bars horizontal by switching the axes. 

## Heatmap

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors. Let's use an example to better understand a heatmap. We'll use another sample dataset from Seaborn, called `flights`, to visualize monthly passenger footfall at an airport over a $12$ year period.


```python
flights_df = sns.load_dataset("flights").pivot(index='month', columns='year', values='passengers')

flights_df
```

**TIP:** The `.pivot()` method is a very powerful tool used to reshape datasets. To learn more about it, visit [HERE](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html)

The `flights_df` is a matrix where the rows are `month` and the columns are `year`. The values show the number of passengers (in thousands) that visited the airport in a specific month of a year. We can use the `sns.heatmap()` function to visualize the footfall at the airport.

```python
plt.title("Number of Passengers (in thousands)")
sns.heatmap(flights_df);
```

By looking at the graph, we can infer two things:
- The footfall at the airport in any given year tends to be the highest around July & August.
- The footfall at the airport in any given month tends to grow year by year.

We can also display the actual values in each block by specifying `annot=True` and using the `cmap` argument to change the color palette.

```python
plt.title("Number of Passengers (in thousands)")
sns.heatmap(data = flights_df, 
            fmt = 'd', # idk the purpose, but maybe something about using when adding annotations
            annot = True, 
            cmap = 'Blues');
```

### Contour Plot

Contour plot uses contours or color-coded regions helps us to _visualize 3 numerical variables in two dimensions_. One variable is represented on the horizontal axis and a second variable is represented on the vertical axis. The third variable is represented by a color gradient and isolines (lines of constant value).

```python
plt.title("Flowers")

sns.kdeplot(x = iris_df.sepal_length, 
            y = iris_df.sepal_width, 
            shade=True); # if true, resembles contour map 
```

## Plotting Multiple Charts in a Grid

Matplotlib and Seaborn also support plotting multiple charts in a grid, using `plt.subplots()`, which returns a set of axes for plotting.

The following is a single grid showing the different types of charts we've created in this tutorial...

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8)) # 2x3 subplots / 16x8in overall size

# SubPlot 1
# Use the axes for plotting
        # axis[x,y]. is the placement of this plot in the 2x3 gird
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')


# SubPlot 2
# Pass the axes into seaborn
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=iris_df.sepal_length, 
                y=iris_df.sepal_width, 
                hue=iris_df.species, 
                s=100, 
                ax=axes[0,1]); # <--- places sns plot in the [0,1] position of the subplot grid

# SubPlot 3
# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True); # <--- stacks the three histograms 
axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);


# SubPlot 4
# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex', 
            data=tips_df, 
            ax=axes[1,0]); # <--- places sns plot in the [1,0] position of the subplot grid


# SubPlot 5
# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, 
            cmap='Blues', 
            ax=axes[1,1]); # <--- places sns plot in the [1,1] position of the subplot grid


# SubPlot 6
# Plot an image using the axes
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2); # <--- correlates to the spacing between subplots
```
